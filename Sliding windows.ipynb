```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
from lesson_functions import *
#matplotlib inline
```

```python
img_files = glob.glob('*.png')
for img_file in img_files:
    img = cv2.imread(img_file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    all_bboxes = []
    
    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, 500],
                          xy_window=(32, 32), 
                          xy_overlap=(0.5, 0.5))
    all_bboxes = all_bboxes + bboxes
    img = draw_boxes(img, bboxes, color=(0, 128, 0), thick=2)

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, 500],
                          xy_window=(64, 64), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(0, 256, 0), thick=2)
    all_bboxes = all_bboxes + bboxes

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, None],
                          xy_window=(96, 96), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(128, 0, 0), thick=2)
    all_bboxes = all_bboxes + bboxes

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, None],
                          xy_window=(128, 128), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(255, 0, 0), thick=2)
    all_bboxes = all_bboxes + bboxes

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, None],
                          xy_window=(160, 160), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(0, 0, 128), thick=2)
    all_bboxes = all_bboxes + bboxes

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, None],
                          xy_window=(192, 192), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(0, 0, 255), thick=2)
    all_bboxes = all_bboxes + bboxes

    bboxes = slide_window(img, 
                          x_start_stop=[None, None],
                          y_start_stop=[img.shape[0]//2+32, None],
                          xy_window=(256, 256), 
                          xy_overlap=(0.5, 0.5))
    img = draw_boxes(img, bboxes, color=(255, 255, 255), thick=2)
    all_bboxes = all_bboxes + bboxes

    print(len(all_bboxes))
    plt.figure(figsize=(16, 12))
    plt.imshow(img)
    plt.show()
```
